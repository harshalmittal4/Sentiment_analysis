```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
def tweet_to_words(raw_tweet):
    letters_only = re.sub("[^a-zA-Z@]", " ",raw_tweet)
    words = letters_only.lower().split()
    stops = set(stopwords.words("english"))
    meaningful_words = [w for w in words if not w in stops and not re.match("^[@]", w) and not re.match("flight",w)]
    return( " ".join( meaningful_words ))
```

```python

Tweet['clean_tweet']=Tweet['text'].apply(lambda x: tweet_to_words(x))

Tweet['sentiment'] = Tweet['airline_sentiment'].apply(lambda x: 0 if x == 'negative' else 1)
 
```

```python
all_text = ' '.join(Tweet['clean_tweet']) #join all clean tweets
words = all_text.split()#words contains the vocab for the entire dataset tweets
```

```python
counts = Counter(words) #dict with keys as words and values as their no. of occurences(0,1,..)
vocab = sorted(counts, key=counts.get, reverse=True)# counts sorted based on values
vocab_to_int = {word: ii for ii, word in enumerate(vocab, 1)}
 
tweet_ints = []
for each in Tweet['clean_tweet']:
    tweet_ints.append([vocab_to_int[word] for word in each.split()])
```

```python
labels = np.array([0 if each == 'negative' else 1 for each in Tweet['airline_sentiment'][:]])
tweet_len = Counter([len(x) for x in tweet_ints])
print("Zero-length reviews: {}".format(tweet_len[0]))
print("Maximum review length: {}".format(max(tweet_len)))

#remove tweets having 0 length and those labels
tweet_idx  = [idx for idx,tweet in enumerate(tweet_ints) if len(tweet) > 0]
labels = labels[tweet_idx]
Tweet = Tweet.iloc[tweet_idx]
tweet_ints = [tweet for tweet in tweet_ints if len(tweet) > 0]
```

```python
seq_len = max(tweet_len)
features = np.zeros((len(tweet_ints), seq_len), dtype=int)
for i, row in enumerate(tweet_ints):
    features[i, -len(row):] = np.array(row)[:seq_len]
```

```python
split_frac = 0.8
split_idx = int(len(features)*0.8)
train_x, val_x = features[:split_idx], features[split_idx:]
train_y, val_y = labels[:split_idx], labels[split_idx:]

test_idx = int(len(val_x)*0.5)
val_x, test_x = val_x[:test_idx], val_x[test_idx:]
val_y, test_y = val_y[:test_idx], val_y[test_idx:]
```

```python
print("Train set x: {}".format(train_x.shape), 
      "Validation set x: {}".format(val_x.shape),
      "Test set x: {}".format(test_x.shape))
```

```python
lstm_size = 256    #cells
lstm_layers = 1
batch_size = 100
learning_rate= 0.001
```

```python
print(len(vocab_to_int))
```

# Graph

```python
n_words= len(vocab_to_int)
graph=tf.Graph()
with graph.as_default():
    inputs_ = tf.placeholder(tf.int32, [None, None], name='inputs')
    labels_ = tf.placeholder(tf.int32, [None, None], name='labels')
    keep_prob = tf.placeholder(tf.float32, name='keep_prob')
    
embed_size= 300 #each word mapped to 300d vector entries of which are learnt during training 
with graph.as_default():
    embedding = tf.Variable(tf.random_uniform((12767, embed_size), -1, 1)) #12767 = n_words +1
    embed = tf.nn.embedding_lookup(embedding, inputs_) # the highest integer given to a word is 12766, it will look at row having index
                                                         # 12766 in embedding, so use 12767 above
    
with graph.as_default():
    lstm= tf.contrib.rnn.BasicLSTMCell(lstm_size)
    drop = tf.contrib.rnn.DropoutWrapper(lstm, output_keep_prob = keep_prob)
    cell = tf.contrib.rnn.MultiRNNCell([drop]*lstm_layers)#stacking multiple lstm layers
    initial_state = cell.zero_state(batch_size,tf.float32)# initial state of zero for each ex. in batch


with graph.as_default():
    outputs, final_state = tf.nn.dynamic_rnn(cell , embed, initial_state= initial_state)

# prediction made from output of last cell
with graph.as_default():
    predictions = tf.contrib.layers.fully_connected(outputs[:,-1],1,activation_fn= tf.sigmoid)
    
    cost= tf.losses.mean_squared_error(labels_,predictions)
    optimizer = tf.train.AdamOptimizer(learning_rate).minimize(cost)
    
```

```python
#accuracy check
with graph.as_default():
    correct_pred= tf.equal(tf.cast(tf.round(predictions) ,tf.int32), labels_)
    accuracy = tf.reduce_mean(tf.cast(correct_pred, tf.float32))
```

```python
#batching (only full batches)
def get_batches(x,y,batch_size=100):
    n_batches = len(x)//batch_size
    x,y= x[:n_batches*batch_size], y[:n_batches*batch_size]
    for ii in range(0,len(x), batch_size):
        yield x[ii:ii+batch_size],y[ii:ii+batch_size]
    
    
    
```

# Training


```python
epochs=10
with graph.as_default():
    saver = tf.train.Saver()
with tf.Session(graph=graph) as sess:
    sess.run(tf.global_variables_initializer())
    iteration =1
    for e in range(epochs):
        state= sess.run(initial_state)
        
        for ii, (x,y) in enumerate(get_batches(train_x,train_y,batch_size),1):
            feed = {inputs_: x,
                    labels_: y[:, None],
                    keep_prob: 0.5,
                    initial_state: state}
            loss, state, _ = sess.run([cost, final_state, optimizer], feed_dict=feed)
            #after each iteration thru this the weigths are updated and new loss and state are calculated
            
            
            if iteration%5==0:
                print("Epoch: {}/{}".format(e, epochs),
                      "Iteration: {}".format(iteration),
                      "Train loss: {:.3f}".format(loss))
                
                
            #print accuracy on validation set after every 25 iterations
            if iteration%25==0:
                val_acc = []
                val_state = sess.run(cell.zero_state(batch_size, tf.float32))
                for x, y in get_batches(val_x, val_y, batch_size):
                    feed = {inputs_: x,
                            labels_: y[:, None],
                            keep_prob: 1,
                            initial_state: val_state}
                    batch_acc, val_state = sess.run([accuracy, final_state], feed_dict=feed)
                    val_acc.append(batch_acc)
                print("Val acc: {:.3f}".format(np.mean(val_acc)))
                
            iteration +=1
    saver.save(sess, "checkpoints/twitter_sentiment.ckpt")
```

# Testing

```python
test_acc=[]
test_pred=[]
with tf.Session(graph=graph) as sess:
    saver.restore(sess,tf.train.latest_checkpoint('checkpoints'))
    test_state = sess.run(cell.zero_state(batch_size,tf.float32))
    for x,y in get_batches(test_x,test_y,batch_size):
        feed ={inputs_:x,
               labels_:y[:,None],
                keep_prob:1,
                initial_state: test_state
               }
        
        batch_acc, test_state= sess.run([accuracy,final_state], feed_dict=feed)
        test_acc.append(batch_acc)
        
        prediction = tf.cast(tf.round(predictions),tf.int32)
        prediction = sess.run(prediction,feed_dict=feed)
        test_pred.append(prediction)
    print("Test accuracy: {:.3f}".format(np.mean(test_acc)))
   
    

    
```

```python
test_pred_flat = (np.array(test_pred)).flatten() # converge test_pred into a single 1d array
print (test_pred_flat.shape)
start_idx = len(train_x) + len(val_x)
end_idx = start_idx + len(test_pred_flat)+1
Tweet.loc[start_idx:end_idx,'predicted_sentiment'] = test_pred_flat


from os import path
from wordcloud import WordCloud
import matplotlib.pyplot as plt

test_tweets = Tweet[start_idx:]
negActualTweets = test_tweets[test_tweets.sentiment!=1]
negPredTweets = test_tweets[test_tweets.predicted_sentiment!=1]

tweetText = ' '.join((negActualTweets['clean_tweet']))

fig = plt.figure( figsize=(20,20))
sub1= fig.add_subplot(2,2,1)
# Generate a word cloud image
wordcloud = WordCloud().generate(tweetText)
plt.imshow(wordcloud, interpolation='bilinear')
plt.title("Negative Sentiment - Actual")
plt.axis("off")

sub2= fig.add_subplot(2,2,2)
tweetText = ' '.join((negPredTweets['clean_tweet']))
wordcloud = WordCloud().generate(tweetText)
plt.title("Negative Sentiment - Prediction")
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
